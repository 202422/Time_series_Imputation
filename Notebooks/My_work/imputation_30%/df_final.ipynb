```python
df_concat = pd.DataFrame(columns = df_3.columns)
for i in range(1,15):
  df = pd.read_csv("/content/Patient_result_RNN_30_part{}.csv".format(i))
  for j in range(0,df.shape[0]):
    df_concat.loc[df_concat.shape[0]] = df.loc[j]
    Pred = lambda p: (df_concat[df_concat["MMRE"] <= p]).shape[0]/df_concat.shape[0]*100
    PRED = Pred(25)
    pred_df.loc[pred_df.shape[0]] = PRED
```
